```python
import pandas as pd
import numpy as np
from sqlalchemy import create_engine
import urllib

# إعداد الاتصال بـ SQL Server باستخدام Windows Authentication
params = urllib.parse.quote_plus(
    "DRIVER={ODBC Driver 17 for SQL Server};"
    "SERVER=DESKTOP-5PAB7M2;"
    "DATABASE=HIV_DB;"
    "Trusted_Connection=yes;"
)

engine = create_engine(f"mssql+pyodbc:///?odbc_connect={params}")

# أسماء الجداول
tables = [
    'art_coverage',
    'art_pediatric',
    'mother_to_child_transmission',
    'no_of_cases_adults_15_49',
    'no_of_deaths',
    'no_of_people_living_with_hiv'
]

dfs = {}

for table in tables:
    df = pd.read_sql(f"SELECT * FROM {table}", engine)
    
    # ملء القيم المفقودة في الأعمدة الرقمية فقط
    for col in df.select_dtypes(include=[np.number]).columns:
        df[col].fillna(df[col].mean(), inplace=True)
    
    dfs[table] = df

# مثال على عرض أول 5 صفوف من جدول
dfs['art_coverage'].head()
```

```python
import pandas as pd
import numpy as np
from sqlalchemy import create_engine

# اتصال بالـ SQL Server
server = 'DESKTOP-5PAB7M2'
database = 'HIV_DB'  # ← غيّر دا لاسم قاعدة البيانات الحقيقي بتاعك
engine = create_engine(f"mssql+pyodbc://@{server}/{database}?trusted_connection=yes&driver=ODBC+Driver+17+for+SQL+Server")

# تحميل كل الجداول
df_coverage = pd.read_sql("SELECT * FROM art_coverage", engine)
df_pediatric = pd.read_sql("SELECT * FROM art_pediatric", engine)
df_mtc = pd.read_sql("SELECT * FROM mother_to_child_transmission", engine)
df_cases = pd.read_sql("SELECT * FROM no_of_cases_adults_15_49", engine)
df_deaths = pd.read_sql("SELECT * FROM no_of_deaths", engine)
df_people = pd.read_sql("SELECT * FROM no_of_people_living_with_hiv", engine)
df_countries = pd.read_sql("SELECT * FROM countries", engine)

# إزالة الأعمدة المتكررة
for df in [df_pediatric, df_mtc, df_cases, df_deaths, df_people, df_countries]:
    if 'WHO Region' in df.columns:
        df.drop(columns=['WHO Region'], inplace=True)

# الدمج النهائي
df = df_coverage.merge(df_pediatric, on='CountryID', suffixes=('_adults', '_pediatric')) \
                .merge(df_mtc, on='CountryID') \
                .merge(df_cases, on='CountryID') \
                .merge(df_deaths, on='CountryID', suffixes=('', '_deaths')) \
                .merge(df_people, on='CountryID') \
                .merge(df_countries, on='CountryID')

# ملء الفراغات بالقيم المتوسطة
for col in df.select_dtypes(include=[np.number]).columns:
    df[col].fillna(df[col].mean(), inplace=True)

# عرض النتائج
df.head()
```
